```python
# Filtering

# Filter for movies only
movies_df = data[data["type"] == "Movie"]
movies_df
```

```python
# Group by release year and count the number of movies released each year
release_counts = movies_df["release_year"].value_counts()
release_counts 
```

```python
# Get the year with the maximum number of movies released
most_released_year = release_counts.idxmax()
most_released_count = release_counts.max()

print(f"The year with the most movies released is {most_released_year} with {most_released_count} movies.")
```
