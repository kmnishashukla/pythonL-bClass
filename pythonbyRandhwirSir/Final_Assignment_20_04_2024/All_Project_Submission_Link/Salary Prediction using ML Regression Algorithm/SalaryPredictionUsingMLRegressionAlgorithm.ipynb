                             In this project the model predicts the salary based on years of experience and plots a graph for it.
                             We import the salary_predict_dataset and analayse the data using different python libraries and trained the regression model.

```python
x = dataset.iloc[:,:1].values # fetching the years of experiance column
x
```

```python
y = dataset.iloc[:,1].values # featching the salary of the dataset
y
```

```python
from sklearn.model_selection import train_test_split # spliting the dataset to train and testing on it
from sklearn.linear_model import LinearRegression

x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=1/3,random_state=0)

regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)

#plotting the graph
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train))
plt.show()
```
