```python
import tkinter as tk
import mysql.connector # pip install mysql-connector-python

#Mysql database connect
def connect_to_database():
    mydb1=mysql.connector.connect(
        host="localhost",
        user ="root",
        passwd="nisha123",
        database="inventory",
        auth_plugin="mysql_native_password"
    )
    return mydb1

# Create function for add a product to the database
def addProduct(name,quantity,price):
    db = connect_to_database()
    cursor = db.cursor()
    query1 = "INSERT INTO products (name,quantity,price) values (%s,%s,%s)"
    cursor.execute(query1,(name,quantity,price))
    db.commit()
    db.close()

# Create function for display all products
def displayProducts():
    db = connect_to_database()
    cursor = db.cursor()
    cursor.execute("select * from products ")
    products = cursor.fetchall()
    for product in products:
        print(product)
    db.close()

# Tkinter gui
def addProductWindow():
    def add_product_to_db():
        name = name_entry.get()
        quantity = int(quantity_entry.get())
        price = int(price_entry.get())
        addProduct(name,quantity,price)
        name_entry.delete(0,tk.END)
        quantity_entry.delete(0,tk.END)
        price_entry.delete(0,tk.END)

    root = tk.Tk()
    root.title("Add product")

    name_label = tk.Label(root,text="Name:")
    name_label.pack()
    name_entry = tk.Entry(root)
    name_entry.pack()

    name_label = tk.Label(root,text="Quantity:")
    name_label.pack()
    name_entry = tk.Entry(root)
    name_entry.pack()

    name_label = tk.Label(root,text="Price:")
    name_label.pack()
    name_entry = tk.Entry(root)
    name_entry.pack()

    addButton = tk.Button(root,text="Add Product",command = add_product_to_db)
    addButton.pack()

    root.mainloop()

addProductWindow()
displayProducts()



```
