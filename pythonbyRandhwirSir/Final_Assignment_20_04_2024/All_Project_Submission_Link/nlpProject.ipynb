```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
def preprocessing_text(text):
 text=text.lower()
 text=re.sub(r'[^a-zA-Z0-9\s]',"",text)
 words=word_tokenize(text)
 stop_words=set(stopwords.words('english'))
 words=[word for word in words if word not in stop_words]
 stemmer=PorterStemmer()
 words=[stemmer.stem(word) for word in words]
 preprocessed_text="".join(words)
 return preprocessed_text
```

```python
from sentence_transformers import SentenceTransformer,util
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def suggest_sections(complaint,new_ds,min_suggestions=5):
    preprocessed_complaint=preprocessing_text(complaint)
    complaint_embedding=model.encode(preprocessed_complaint)
    section_embeddings=model.encode(new_ds['Combo'].tolist())
    similarities=util.pytorch_cos_sim(complaint_embedding,section_embeddings)[0]
    similarity_threshold=0.3
    relevant_indices=[]
    while len(relevant_indices)<min_suggestions and similarity_threshold>0:
        relevant_indices=[i for i,sim in enumerate(similarities) if sim>similarity_threshold]
        similarity_threshold-=0.05
        sorted_indices=sorted(relevant_indices,key=lambda i: similarities[i],reverse=True)
        suggestions=new_ds.iloc[sorted_indices][['Description','Offense','Punishment','Cognizable','Bailable','Court']].to_dict(orient='records')
        return suggestions
```

```python
complaint=input("Enter crime description")
suggested_sections=suggest_sections(complaint,new_ds)
print(suggested_sections)
if suggested_sections:
    print("Suggested section")
    for s in suggested_sections:
        print(f"Description :{s['Description']}")
        print(f"Offense:{s['Offense']}")
        print(f"Punishment {s['Punishment']}")
else:
    print("No relevant section found")
```
