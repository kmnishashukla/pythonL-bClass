Scatter plot:

    Bivariate Analysis:To study 2 columns we can make a use of scatter plot

    Numerical vs Numerical

    Use case: Finding Correlation

    example:Plotting study time(hr) on the x-axis against exam scores(out of 100)on y-axis for a class of students.

```python
import matplotlib.pyplot as plt
import numpy
numpy.random.seed(0)
study_time=numpy.random.randint(1,6,50)
exam_score=50+10*study_time+numpy.random.normal(0,5,50)
print(study_time)
print(exam_score)
```

```python
plt.scatter(study_time,exam_score)
plt.xlabel("Study time")
plt.ylabel("Score")
plt.title("Student performance based on time")
plt.show()
```

Line Plot

        Bivariate analysis:Both the columns should be numeric
        or
        One is numeric and other is categorical

        Use case: time series 

```python
import numpy as np
import pandas as pd
dates=pd.date_range(start="2024-01-01",periods=30)
price=100+np.cumsum(np.random.normal(0,1,30))
print(dates)
print(price)
```

```python
import numpy as np
import pandas as pd
dates=pd.date_range(start="2024-01-01",periods=30)
price=100+np.cumsum(np.random.normal(0,1,30))
print(dates)
print(price)
plt.plot(dates,price)
plt.xlabel("Dates")
plt.ylabel("Price")
plt.title("Product price in jan month")
plt.show()
```

Bar chart:
Bivariate analysis:Numeric vs categorical

Use case:Aggregate analysis of group




```python
product=["A","B","C"]
sales=[1000,900,1200]
plt.bar(product,sales,color=["Blue","Red","Green"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product sale")
plt.show()
```
